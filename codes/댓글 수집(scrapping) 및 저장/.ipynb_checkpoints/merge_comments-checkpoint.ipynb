```python
import pandas as pd
```

```python
df_one = pd.read_csv('../../datasets/HyeIn.csv')
df_one[:5]
```

```python
df_two = pd.read_csv('../../datasets/bwh.csv')
df_two[:5]
```

### DB Concat

```python
# 필요없는 column 삭제
df_one.drop(columns=['Unnamed: 0','_id'], axis=1, inplace=True)
```

```python
df_one[:5]
```

```python
df_two.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df_two[:5]
```

```python
# column 위치 교체
df_two = df_two[['writer','date','review']]
df_two[:5]
```

```python
# db 병합
total_db = pd.concat([df_one, df_two])
total_db[:5]
```

```python
# 저장
total_db.to_csv('../../datasets/total_db.csv')
```
